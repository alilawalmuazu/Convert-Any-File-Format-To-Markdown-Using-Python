```python
import nbformat
from nbformat.v4 import new_notebook, new_code_cell
import json
import csv
import PyPDF2
from docx import Document
import pandas as pd

def convert_csv_to_markdown(csv_file_path):
    # Read CSV file
    with open(csv_file_path, 'r') as csv_file:
        reader = csv.reader(csv_file)
        data = list(reader)

    # Convert data to Markdown
    markdown = ''
    for row in data:
        markdown += ' | '.join(row) + '\n'
    
    return markdown

def convert_pdf_to_markdown(pdf_file_path):
    # Read PDF file
    with open(pdf_file_path, 'rb') as pdf_file:
        reader = PyPDF2.PdfFileReader(pdf_file)
        text = ''
        for page_num in range(reader.numPages):
            page = reader.getPage(page_num)
            text += page.extractText()
    
    return text

def convert_word_to_markdown(word_file_path):
    # Read Word file
    doc = Document(word_file_path)
    
    # Convert paragraphs to Markdown
    markdown = ''
    for paragraph in doc.paragraphs:
        markdown += paragraph.text + '\n'
    
    return markdown

def convert_excel_to_markdown(excel_file_path):
    # Read Excel file
    df = pd.read_excel(excel_file_path)
    
    # Convert DataFrame to Markdown
    markdown = df.to_markdown(index=False)
    
    return markdown

def convert_txt_to_markdown(txt_file_path):
    # Read TXT file
    with open(txt_file_path, 'r') as txt_file:
        text = txt_file.read()
    
    return text

# Example usage:
csv_file_path = 'example.csv'
pdf_file_path = 'example.pdf'
word_file_path = 'example.docx'
excel_file_path = 'example.xlsx'
txt_file_path = 'example.txt'

csv_markdown = convert_csv_to_markdown(csv_file_path)
pdf_markdown = convert_pdf_to_markdown(pdf_file_path)
word_markdown = convert_word_to_markdown(word_file_path)
excel_markdown = convert_excel_to_markdown(excel_file_path)
txt_markdown = convert_txt_to_markdown(txt_file_path)

# Now you can write these Markdown strings to files or further process them as needed.
```
